```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_pickle('Kinages_Prot.pkl')
```

```python
cols=['kepid','Prot','Prot_err','source_id','ra','ra_error','dec',
     'dec_error','all_vz','vz_err_all','vel_dis','kin_age','kin_age_err_AVR',
     'kin_age_err','kin_age_err_t','Ro','teff','abs_G']
data['teff']=data['color_teffs']
```

```python
data['kin_age_err_t']=np.sqrt((data['kin_age_err_AVR'])**2+(data['kin_age_err'])**2)
data['source_id']=np.array(data['source_id'],dtype=int)
data[cols].to_csv('Kinages_Prot.csv', index = False)
```
